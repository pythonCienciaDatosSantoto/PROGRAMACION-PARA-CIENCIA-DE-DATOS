Desde su creación, se ha convertido en un proyecto de código abierto con una comunidad global de contribuidores que lo mejoran constantemente.

* **1.3. La Estructura Fundamental: `pd.Series`**

Una **Serie** es el bloque de construcción básico de Pandas. Es un arreglo de una dimensión que puede contener cualquier tipo de dato (números, texto, fechas, etc.). Su característica principal es que cada elemento tiene una etiqueta o **índice**.

#### **Creación de Series**

```python
import pandas as pd
import numpy as np # Es una buena práctica importarlo también

# 1. A partir de una lista (Pandas crea un índice numérico por defecto)
calificaciones = pd.Series([9.5, 8.0, 10.0, 7.5])
print("Serie a partir de una lista:")
print(calificaciones)
print("-" * 30)
```

```python
# 2. Con un índice personalizado (mucho más descriptivo)
calificaciones_materias = pd.Series(
    [9.5, 8.0, 10.0],
    index=['Cálculo', 'Álgebra', 'Programación']
)
print("Serie con índice personalizado:")
print(calificaciones_materias)
print("-" * 30)
```

```python
# 3. A partir de un diccionario (las claves se convierten en el índice)
poblacion_dict = {'Bogotá': 7900000, 'Medellín': 2500000, 'Cali': 2200000}
poblacion_ciudades = pd.Series(poblacion_dict)
print("Serie a partir de un diccionario:")
print(poblacion_ciudades)
```

#### **Atributos y Operaciones Esenciales**

Una vez que tienes una Serie, puedes explorarla y realizar cálculos sobre ella.

```python
# Usando la serie de población
print(f"Índice de la serie: {poblacion_ciudades.index}")
print(f"Valores de la serie: {poblacion_ciudades.values}")
print(f"Tipo de datos: {poblacion_ciudades.dtype}")
print("-" * 30)

# Accediendo a datos
print(f"Población de Medellín: {poblacion_ciudades['Medellín']}")
print("-" * 30)

# Operaciones estadísticas (¡aquí está la magia!)
print(f"Población total: {poblacion_ciudades.sum():,}")
print(f"Población promedio: {poblacion_ciudades.mean():,.2f}")
print(f"Ciudad más poblada tiene: {poblacion_ciudades.max():,} habitantes")
print(f"Ciudad menos poblada tiene: {poblacion_ciudades.min():,} habitantes")
print(f"Mediana de la población: {poblacion_ciudades.median():,.2f}")
print("-" * 30)

# Operaciones vectorizadas (se aplican a todos los elementos)
poblacion_millones = poblacion_ciudades / 1_000_000
print("Población en millones:")
print(poblacion_millones)
```

* **1.4. Visualización Rápida con `.plot()`**

Una imagen vale más que mil números. Pandas se integra con Matplotlib para que puedas crear gráficos directamente desde tus Series con un solo comando.

```python
import matplotlib.pyplot as plt

# Gráfico de barras de la población
poblacion_ciudades.plot(
    kind='bar',
    title='Población de Ciudades Principales de Colombia',
    xlabel='Ciudad',
    ylabel='Número de Habitantes (en millones)',
    rot=45, # Rota las etiquetas para que no se superpongan
    grid=True,
    color=['#FFCD00', '#003893', '#C8102E'] # Colores de la bandera
)
plt.tight_layout() # Ajusta el gráfico para que todo quepa
plt.show()

# Gráfico de pastel (pie chart)
poblacion_ciudades.plot(
    kind='pie',
    autopct='%1.1f%%', # Formato para mostrar porcentajes
    title='Proporción de Población por Ciudad'
)
plt.ylabel('') # Quita la etiqueta del eje y que no aporta nada en un pie chart
plt.show()
```
