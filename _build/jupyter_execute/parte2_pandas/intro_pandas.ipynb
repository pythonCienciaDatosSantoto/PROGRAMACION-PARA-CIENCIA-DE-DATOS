* **1.3. La Estructura Fundamental: `pd.Series`**

Una **Serie** es el bloque de construcci칩n b치sico de Pandas. Es un arreglo de una dimensi칩n que puede contener cualquier tipo de dato (n칰meros, texto, fechas, etc.). Su caracter칤stica principal es que cada elemento tiene una etiqueta o **칤ndice**.

#### **Creaci칩n de Series**

```python
import pandas as pd
import numpy as np # Es una buena pr치ctica importarlo tambi칠n

# 1. A partir de una lista (Pandas crea un 칤ndice num칠rico por defecto)
calificaciones = pd.Series([9.5, 8.0, 10.0, 7.5])
print("Serie a partir de una lista:")
print(calificaciones)
print("-" * 30)
```

```python
# 2. Con un 칤ndice personalizado (mucho m치s descriptivo)
calificaciones_materias = pd.Series(
    [9.5, 8.0, 10.0],
    index=['C치lculo', '츼lgebra', 'Programaci칩n']
)
print("Serie con 칤ndice personalizado:")
print(calificaciones_materias)
print("-" * 30)
```

```python
# 3. A partir de un diccionario (las claves se convierten en el 칤ndice)
poblacion_dict = {'Bogot치': 7900000, 'Medell칤n': 2500000, 'Cali': 2200000}
poblacion_ciudades = pd.Series(poblacion_dict)
print("Serie a partir de un diccionario:")
print(poblacion_ciudades)
```

#### **Atributos y Operaciones Esenciales**

Una vez que tienes una Serie, puedes explorarla y realizar c치lculos sobre ella.

```python
# Usando la serie de poblaci칩n
print(f"칈ndice de la serie: {poblacion_ciudades.index}")
print(f"Valores de la serie: {poblacion_ciudades.values}")
print(f"Tipo de datos: {poblacion_ciudades.dtype}")
print("-" * 30)

# Accediendo a datos
print(f"Poblaci칩n de Medell칤n: {poblacion_ciudades['Medell칤n']}")
print("-" * 30)

# Operaciones estad칤sticas (춰aqu칤 est치 la magia!)
print(f"Poblaci칩n total: {poblacion_ciudades.sum():,}")
print(f"Poblaci칩n promedio: {poblacion_ciudades.mean():,.2f}")
print(f"Ciudad m치s poblada tiene: {poblacion_ciudades.max():,} habitantes")
print(f"Ciudad menos poblada tiene: {poblacion_ciudades.min():,} habitantes")
print(f"Mediana de la poblaci칩n: {poblacion_ciudades.median():,.2f}")
print("-" * 30)

# Operaciones vectorizadas (se aplican a todos los elementos)
poblacion_millones = poblacion_ciudades / 1_000_000
print("Poblaci칩n en millones:")
print(poblacion_millones)
```

* **1.4. Visualizaci칩n R치pida con `.plot()`**

Una imagen vale m치s que mil n칰meros. Pandas se integra con Matplotlib para que puedas crear gr치ficos directamente desde tus Series con un solo comando.

```python
import matplotlib.pyplot as plt

# Gr치fico de barras de la poblaci칩n
poblacion_ciudades.plot(
    kind='bar',
    title='Poblaci칩n de Ciudades Principales de Colombia',
    xlabel='Ciudad',
    ylabel='N칰mero de Habitantes (en millones)',
    rot=45, # Rota las etiquetas para que no se superpongan
    grid=True,
    color=['#FFCD00', '#003893', '#C8102E'] # Colores de la bandera
)
plt.tight_layout() # Ajusta el gr치fico para que todo quepa
plt.show()

# Gr치fico de pastel (pie chart)
poblacion_ciudades.plot(
    kind='pie',
    autopct='%1.1f%%', # Formato para mostrar porcentajes
    title='Proporci칩n de Poblaci칩n por Ciudad'
)
plt.ylabel('') # Quita la etiqueta del eje y que no aporta nada en un pie chart
plt.show()
```
